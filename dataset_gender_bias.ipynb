```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/queries-documents_Sex & Relationship.csv")
```

```python
df
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Filter the DataFrame for content_gender 'F'
df_f = df[df['content_gender'] == 'F']
text_f = " ".join(df_f['query'].astype(str)) + " " + " ".join(df_f['doc_title'].astype(str)) + " " + " ".join(df_f['document'].astype(str))
wordcloud_f = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = STOPWORDS,
                min_font_size = 10).generate(text_f)

# Filter the DataFrame for content_gender 'M'
df_m = df[df['content_gender'] == 'M']
text_m = " ".join(df_m['query'].astype(str)) + " " + " ".join(df_m['doc_title'].astype(str)) + " " + " ".join(df_m['document'].astype(str))
wordcloud_m = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = STOPWORDS,
                min_font_size = 10).generate(text_m)

# Plot the word clouds
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud_f)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.title("Wordcloud for content_gender = F")
plt.show()

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud_m)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.title("Wordcloud for content_gender = M")
plt.show()
```
